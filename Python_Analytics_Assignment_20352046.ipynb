**1) CREATE A DATAFRAME**



A Data Frame is a two-dimension collection of data. It is a data structure where data is stored in tabular form. Datasets are arranged in rows and columns; we can store multiple datasets in the data frame. We can perform various arithmetic operations, such as adding column/row selection and columns/rows in the data frame.





>  a) DataFrame using list.

Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is generally the most commonly used pandas object.


```python
# Import pandas library
import pandas as pd
  
# initialize list of lists
data = [['Person of Interest', 95], ['Breaking Bad', 85], ['Prison Break', 80],['Blindspot',75]]
  
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['TV SHOW', 'SCORE'])
  
# print dataframe.
df
```

zip() function creates the objects and that can be used to produce single item at a time. This function can create pandans DataFrames by merging two lists.

```python
import pandas as pd 
    
# List1 
TV_SHOW = ['Person of Interest', 'Breaking Bad', 'Prison Break', 'Blindspot'] 
    
# List2 
SCORE = [95, 85, 80, 75] 
    
# get the list of tuples from two lists. 
# and merge them by using zip(). 
list_of_tuples = list(zip(TV_SHOW, SCORE)) 
    
# Assign data to tuples. 
list_of_tuples  
  
  
# Converting lists of tuples into 
# pandas Dataframe. 
df = pd.DataFrame(list_of_tuples,
                  columns = ['TV_SHOW', 'SCORE']) 
     
# Print data. 
df 
```

**4. COMBINING AND MERGING DATA FRAME**





```python
import pandas as pd
```



> a) Concatenating DataFrame







> 4.a) 1 - using concat only.



```python
import pandas as pd 
series1 = {'Series':['CHERNOBYL', 'THE FLASH', 'ARROW', 'SWAMP THING'], 
        'Year':[2014, 2010, 2012, 2015], 
        'Genre':['Documentary','Action', 'Adventure', 'Fantasy']} 
   
# Define dictionary 
series2 = {'Series':['IRON FIST', 'TEEN WOLF', 'DAREDEVIL', 'JESSICA JONES'], 
        'Year':[2018,2009, 2011,2017], 
        'Genre':['Adventure', 'Drama', 'Action', 'Thriller']} 

 
# Convert the dictionarys into DataFrames  
df1= pd.DataFrame(series1,index=[0, 1, 2, 3])
df2 = pd.DataFrame(series2, index=[4, 5, 6, 7])
frames = [df1, df2]
final_1 = pd.concat(frames)
final_1
```



> 4.a) 2 - Setting axis along with concat() to get intersection of dataframe.



```python
series3 = {'Series':['IRON FIST', 'TEEN WOLF', 'DAREDEVIL', 'JESSICA JONES'], 
        'Year':[2018,2009, 2011,2017],  'Genre':['Adventure', 'Drama', 'Action', 'Thriller']} 
df3 = pd.DataFrame(series3, index= [2,3,5,7])
end_2 = pd.concat([df1, df3], axis=1, join='inner')
end_2
```



> b) Merging DataFrame.



Merging two dataset can be done in dataframes with common column name and with different column name.


>   4.b) 1 -  Merge Two DataFrames on a Key.




```python
import pandas as pd
left = pd.DataFrame({
   'No':[1,2,3,4,5],
   'Car': ['Creta', 'Swift', 'Ecosport', 'Innova', 'i20'],
   'type':['Diesel','Petrol','Diesel','Petrol','Electric']})
right = pd.DataFrame(
   {'No':[1,2,3,4,5],
   'Model': ['Suv', 'Offroad', 'Suv', 'Offroad', 'Compact'],
   'type':['Economy','Smooth','Economy','Smooth','Best']})
pd.merge(left,right,on='No')
```



>   4.b) 2 -  Merge Two DataFrames on multiple Keys.



```python
import pandas as pd
left = pd.DataFrame({
   'No':[1,2,3,4,5],
   'Car': ['Creta', 'Swift', 'Ecosport', 'Innova', 'i20'],
   'type':['Diesel','Petrol','Diesel','Petrol','Electric']})
right = pd.DataFrame(
   {'No':[1,2,3,4,5],
   'Model': ['Suv', 'Offroad', 'Suv', 'Offroad', 'Compact'],
   'type':['Diesel','Petrol','Diesel','Petrol','Electric']})
pd.merge(left,right,on=['No','type'])
```

  

> 4.b) 3 -  Merge Using 'how' Argument



```python
# Left Join
pd.merge(left,right,on='type', how='left')
```

```python
#right join
pd.merge(left, right, on='type', how='right')
```

```python
#outer join
pd.merge(left, right, on='type', how='outer')
```

```python
#inner join
pd.merge(left, right, on='type', how='inner')
```

**8. PIVOTING OPERATIONS IN DATAFRAME.**


> Pivoting dataframes reshaping datframes. In some scenario we might need to have column values as column name or index in those scenarios pivotting comes into the picture. Panda provides two methods for pivotting.


*   pivot() - Returns ReshapedDataFrame
*   pivot_table() - Returns DataFrame





```python
import pandas as pd 

data = {'Room No':[802,805,808,811,815],
        'Guest':['Prasath','Vishwa','Raja','Mani','Keerthi'],
        'Rent':[2000,3000,3500,3800,7000]}                

df = pd.DataFrame(data)
df
```

8) a - Pivot()


> pivot() is an extremely powerfull function that can be used to reshape dataframe based on colums values. pivot() is called using DataFrame object and it returns ReshapedDataFrame object. pivot() contains three parameters they are:


*   **index**- Single/List of columns to make new frame's index
*   **columns** - Single/List of columns to make new frame's column
*    **values** - Sinlge/List of columns to populate new frame

**Syntax :**

Pivot(index,columns,values)

```python
df.pivot(index ='Room No',columns = 'Guest',values = 'Rent' )
```

**8.2 pivot_table()**

Like pivot() pivot_table() is also used to reshape dataframe but pivot_table returns DataFrame which look like excel style pivot table. And pivot_table is called using pandas object. pivot_table provides many parameter some of the important params are:

*   **data** - DataFrame object which holds data
*   **index** - Single/List of columns to make new frame's index
*   **columns** - Single/List of columns to make new frame's column
*   **values** - Sinlge/List of columns to populate new frame

**Syntax :**

pivot_table(data,index,columns,values)

```python
df2 = pd.pivot_table(data = df,index = 'Guest',columns = 'Room No',values ='Rent')
df2
```

Splits each string with the given pattern. Strings are split and the new elements after the performed split operation, are stored in a list.
